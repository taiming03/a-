```python
import cv2, numpy as up
from google.colab.patches import cv2_imshow
defect = 0

img = cv2.imread('ok.jpg')
# cv2_imshow(img)
size = img.shape
# print("照片原尺寸=",size)
# print(f'size[0]={size[0]}\nsize[1]={size[1]}')

imgr = cv2.resize(img, (450, 250))
# cv2_imshow(imgr)
size = imgr.shape
# print("縮小後尺寸=",size)
# print(f'size[0]={size[0]}\nsize[1]={size[1]}')

gray = cv2.cvtColor(imgr, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)

y = cv2.Sobel(gray, cv2.CV_16S, 0, 1)
absY = cv2.convertScaleAbs(y)
cv2_imshow(absY)
print(absY)

ret, thresh = cv2.threshold(absY,100,255,cv2.THRESH_BINARY)
print("ret=",ret)
cv2_imshow(thresh)
```
